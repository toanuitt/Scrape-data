```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
url_pre = "https://www.realtor.com/realestateandhomes-search/"
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
#scrape data
records = []
count=0
for row in soup.find_all('div', class_=re.compile(r'Cardstyles__StyledCard')):
        count = count+1
        if count % 2 ==1:
            try:
                author = row.find("span",class_=re.compile(r'BrokerTitle_titleText')).get_text()
            except:
                author = None
        else:
            try:
                label_span = row.find("span", {"data-testid": "label-new"}).text
            except:
                label_span = None
            try:
                construction = row.find("span", {"data-testid": "label-new-construction"}).text
            except:
                construction =  None
            try:
                sale = row.find("div", class_=re.compile(r'base__StyledType')).get_text()
                sale = "On"
            except:
                sale = "Off"
            try:
                price_text = row.find("div", class_=re.compile(r'Pricestyles__StyledPrice'), attrs={"data-testid": "card-price"}).text
                price = re.sub("[^\d]", "", price_text)
            except:
                price = 0
            try:
                bed_text = row.find("li", {"data-testid": "property-meta-beds"}).get_text()
                bed =  int(re.search(r'\d+', bed_text).group())
            except:
                bed = 0
            try:
                bath_text = row.find("li", {"data-testid": "property-meta-baths"}).get_text()
                bath =  int(re.search(r'\d+', bath_text).group())
            except:
                bath = 0
            try:
                sqft_text = row.find("li", {"data-testid": "property-meta-sqft"}).get_text()
                sqft =  int(re.search(r'\d+', sqft_text).group())
            except:
                sqft = 0
            try:
                lot_size_text = row.find("li", {"data-testid": "property-meta-lot-size"}).get_text()
                lot_size =  int(re.search(r'\d+', lot_size_text).group())
            except:
                lot_size = 0
            try:
                address1_div = row.find("div", {"data-testid": "card-address-1"})
                address2_div = row.find("div", {"data-testid": "card-address-2"})
              # Get the text content of each address
                address1 = address1_div.text.strip()
                address2 = address2_div.text.strip()
            except:
                address1 = None
                address2 = None  
            if author != None:
                records.append({
                         'label' : label_span,
                         'construction' : construction,
                         'sale' : sale,
                         'price' : price,
                         'bed' : bed,
                         'bath' : bath,
                         'sqft' : sqft,
                         'lot size' : lot_size,
                         'address1' : address1_div,
                         'address2' : address2_div
                      
                    })
            for image in row.find_all('picture'):
                src_value = image.find('img')['src']
                try:
                    image_content = requests.get(src_value).content
                    image_file = io.BytesIO(image_content)
                    image = Image.open(image_file)
                    file_path= "D:/HTML/scrapingdata/Realtor/img/" +address1+address2+".jpg" 
                    with open(file_path,"wb") as f:  
                        image.save(f,"JPEG")
                except:                 
                    continue

            
        
       
       
        
        
```

```python
driver.quit()
```

```python
df_missing = pd.DataFrame.from_dict(records)
df_missing
```

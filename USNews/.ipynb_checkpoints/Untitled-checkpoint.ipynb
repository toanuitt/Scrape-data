```python
from bs4 import BeautifulSoup
import json
import pandas as pd
from selenium import webdriver
import urllib.parse
import re
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Provide the correct path to the chromedriver executable
PATH = r"C:\Program Files (x86)\chromedriver.exe"

# Initialize the Chrome webdriver
driver = webdriver.Chrome(PATH)

# URL for greatschools pre_url and post_url (with state/city in between)
url= 'https://www.usnews.com/best-colleges/search?_sort=rank&_sortDirection=asc'
driver.get(url)
```

```python
scroll_count = 20  # Adjust as needed
for _ in range(scroll_count):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
```

```python
records = []
count=0

for row in soup.find_all('div', class_=re.compile(r'Card__TextContainer')):
        school = row.find('h3',class_=re.compile(r'Heading')).get_text()
        address = row.find('p',class_=re.compile(r'Paragraph')).get_text()
    
        ranking = row.find('a',class_=re.compile(r'RankList__RankLink')).get_text()
        rank_setup = ranking.split(" ", 1)  
        rank = rank_setup[0].replace("#", "").replace("in", "")
        type_universities = rank_setup[1]
    
        tuition_and_fees = row.find('dt', text='Tuition and Fees').find_next('dd').text.strip()
    
        parts = row.find('dt', text='Undergraduate Enrollment').find_next('dd').text.strip().split('(')
        number_part = parts[0].strip() 
        
        
        records.append({
                        'School': school,
                        'Ranking': rank,
                        'Type university': type_universities,
                        'Address': address,
                        'Fee':tuition_and_fees,
                        'Total Students Enrolled': number_part,
                     
                      
                    })
        print(school," ",count)
    
```

```python
df_missing = pd.DataFrame.from_dict(records)
```
